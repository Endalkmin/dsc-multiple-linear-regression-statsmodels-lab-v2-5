## Step 2: Build a Simple Linear Regression Model

Set the dependent variable (`y`) to be the `SalePrice`, then choose one of the features shown in the subset above to be the baseline independent variable (`X`).

Build a linear regression using StatsModels, describe the overall model performance, and interpret its coefficients.

````python
# Your code here - define y and baseline X
y = ames_subset["SalePrice"] 
X_baseline = ames_subset["1stFlrSF"] 
````

````python
# Your code here - import StatsModels, fit baseline model, display results
baseline_model = sm.OLS(y, sm.add_constant(X_baseline)) 
baseline_results = baseline_model.fit() 

print(baseline_results.summary())
````

### My written answer here: 

    Linear Regression line:
$$ \hat{SalePrice} = 36170 + 124.5006 1stFlrSF $$ 
- The model is statistically significant overall, with an F-statistic p-value well below 0.05
- The model explains about 36.7% of the variance in SalePrice
- The model coefficients (const and 1stFlrSF) are both statistically significant, with t-statistic p-values well below 0.05
- If 1stFlrSF(First Floor square feet) is 0 sq ft, we would expect SalePrice to be about 36170
- For each increase of 1 sq ft in 1stFlrSF, we see an associated increase in SalePrice of about 124.4  


## Step 3: Build a Multiple Linear Regression Model

For this model, use **all of** the features in `ames_subset`.

````python
# Your code here - define X 
X_all = ames_subset.drop('SalePrice', axis=1)
````

````python
# Your code here - fit model and display results 
second_model = sm.OLS(y, sm.add_constant(X_all)) 
second_results = second_model.fit() 

print(second_results.summary())
````

````python
# Your written answer here - interpret model results. Does this model seem better than the previous one?
````

### My Answer here:
    Model with all of the features in ames_subset has Linear Regression line:

$$ \hat{SalePrice} = -14310 + 0.2841 LotArea +  60.2866 1stFlrSF + 80.6061 GrLivArea $$

* The model is statistically significant overall, with an F-statistic p-value well below 0.05
* The model explains about 56.5% of the variance in SalePrice
* Increasing the predictors from 1 to 3 increased R-Squared by 19.8% (from 36.7 to 56.5)
* `const`, `1stFlrSF` and `GrLivArea` have p-values below 0.05 and are therefore statistically significant
* `LotArea` has p-value above 0.05. Thus are not statistically significant at an alpha of 0.05. That means it has no effect on SalePrice
